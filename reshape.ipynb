```python
# Import Dependencies
import pandas as pd
```

```python
hurricanes_csv = "state-interactive-2018_v2.csv"
hurricane_df = pd.read_csv(hurricanes_csv)

hurricane_df.head()
```

```python
hurricane_df = hurricane_df.set_index("state")
hurricane_df = hurricane_df.replace('%', '', regex=True)
```

```python
hurricane_df["aff_index_20"] = hurricane_df["aff_index_20"] * 100
hurricane_df["aff_index_35"] = hurricane_df["aff_index_35"] * 100
hurricane_df.head()
```

```python
hurricane_df.to_csv("state-interactive-2018_v2_clean.csv")
```
